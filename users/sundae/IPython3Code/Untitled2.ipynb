```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
f = np.logspace(-1,8,100)
w = 2*np.pi*f
```

```python
def f1(R1,R2,C1):
    out = - R2/(R1*(1j*w*C1*R2+1))
    return out
```

```python
r1 = 1e3
r2 = 1e3
c1 = 390e-12
out1 = f1(r1, r2, c1)
out2 = f1(1e3,1e3,1e-7)
plt.loglog(f,np.abs(out1),label='1')
plt.loglog(f,np.abs(out2),label='2')
plt.legend()
```

```python
plt.loglog(f,np.abs(out1*out2),label='3')
plt.legend()
```
