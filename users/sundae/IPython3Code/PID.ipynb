```python
from sympy import *
from numpy import logspace, pi, angle
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
P, I, D = symbols('P, I, D')
```

```python
def out(f,P=1, I=1, D=1):
    out = -P*(1+I/(1j*2*pi*f)+1j*D*2*pi*f)
    return out
abs(out(1000, P=1, I=0.01, D=0e-6))
```

```python
f = logspace(-1,5,100)
y1 = out(f, P=1, I=100, D=0.0001)
# y2 = out(f, P=1, I=1, D=1e-5)
plt.loglog(f,abs(y1), label='Amp 1');
# plt.loglog(f,abs(y2), label='Amp 2');
plt.legend()
plt.figure()
plt.plot(f,angle(y1,deg=True), label='phase 1')
# plt.plot(f,angle(y2,deg=True), label='phase 1')
plt.legend()
plt.xscale('log')
```
